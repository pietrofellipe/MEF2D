```python
import matplotlib.pyplot as plt

# Número de elementos das malhas
n_elementos = [222, 526, 1555, 2753, 8635, 19165]

# Valores da energia obtidos
energia = [0.218804, 0.22545, 0.228809, 0.229937, 0.230968, 0.231663]

plt.figure(figsize=(8,5))

# Gráfico da energia vs elementos
plt.plot(n_elementos, energia, marker='o', linestyle='-', linewidth=2, markersize=7, label="Energia magnética")

# Eixos em escala log para número de elementos (ajuda a visualizar refinamento)
plt.xscale('log')

# Rótulos e título
plt.xlabel("Número de elementos da malha (escala log)")
plt.ylabel("Energia magnética (J)")
plt.title("Convergência da energia magnética com refinamento da malha")
plt.grid(True, which="both", linestyle="--", linewidth=0.6)
plt.legend()

# Mostrar o gráfico
plt.show()
```
