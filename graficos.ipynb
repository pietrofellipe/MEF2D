```python
import matplotlib.pyplot as plt

# Dados
raios = [0.3967, 0.4687, 0.5408, 0.6129, 0.7211, 1.081, 1.42, 2.16]  # mm
energias = [77.5664, 83.5677, 86.9939, 89.2071, 91.3315, 94.456, 95.5331, 96.4006]

# Plot
plt.figure(figsize=(8,5))
plt.plot(raios, energias, marker='o', linestyle='-', linewidth=1.5, label="Energia magntica")
plt.xlabel("Raio (mm)")
plt.ylabel("Energia (J)")
plt.title("Energia magntica em funo do raio da fronteira externa")
plt.grid(True)

# Destacar o ltimo ponto
plt.scatter(raios[-1], energias[-1], color='red', zorder=5)  # ponto em vermelho
plt.annotate(f"{energias[-1]:.2f}",
             (raios[-1], energias[-1]),
             textcoords="offset points",
             xytext=(-20,-15),
             ha='center',
             color='red')

# Mostrar grfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Nmero de elementos das malhas
n_elementos = [222, 526, 1555, 2753, 8635, 19165]

# Valores da energia obtidos
energia = [0.218804, 0.22545, 0.228809, 0.229937, 0.230968, 0.231663]

plt.figure(figsize=(8,5))

# Grfico da energia vs elementos
plt.plot(n_elementos, energia, marker='o', linestyle='-', linewidth=2, markersize=7, label="Energia magntica")

# Destacar o ponto de maior energia
max_energia = max(energia)
max_idx = energia.index(max_energia)
plt.scatter(n_elementos[max_idx], max_energia, color='red', zorder=5)
plt.annotate(f"{max_energia:.6f}",
             (n_elementos[max_idx], max_energia),
             textcoords="offset points",
             xytext=(-10,-20),
             ha='center',
             color='red')

# Eixos em escala log para nmero de elementos
plt.xscale('log')

# Rtulos e ttulo
plt.xlabel("Nmero de elementos da malha (escala log)")
plt.ylabel("Energia magntica (J)")
plt.title("Convergncia da energia magntica com refinamento da malha")
plt.grid(True, which="both", linestyle="--", linewidth=0.6)
plt.legend()

# Mostrar o grfico
plt.show()
```
